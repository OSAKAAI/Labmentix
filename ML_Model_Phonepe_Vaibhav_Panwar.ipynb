````python
path = "/content/pulse/data/aggregated/transaction/country/india/state"
````

````python
import os
import json
import pandas as pd

data = []

for state in os.listdir(path):
    for year in os.listdir(f"{path}/{state}"):
        for file in os.listdir(f"{path}/{state}/{year}"):
            with open(f"{path}/{state}/{year}/{file}") as f:
                d = json.load(f)

                for i in d['data']['transactionData']:
                    data.append({
                        "State": state,
                        "Year": int(year),
                        "Quarter": int(file.strip('.json')),
                        "Transaction_type": i['name'],
                        "Transaction_count": i['paymentInstruments'][0]['count'],
                        "Transaction_amount": i['paymentInstruments'][0]['amount']
                    })

df = pd.DataFrame(data)
````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Make a copy of dataset
df_clean = df.copy()

# Standardize State names
df_clean["State"] = df_clean["State"].str.replace("-", " ").str.title()

# Sort data for time-based operations
df_clean = df_clean.sort_values(by=["State", "Year", "Quarter"])

# Create Growth Feature (important for ML later)
df_clean["Growth"] = df_clean.groupby("State")["Transaction_amount"].pct_change()

# Handle missing values created by pct_change
df_clean["Growth"] = df_clean["Growth"].fillna(0)

# Create Target Variable (for ML classification)
df_clean["Target"] = df_clean["Growth"].apply(lambda x: 1 if x > 0.15 else 0)

# Final check
df_clean.head()
````

### What all manipulations have you done and insights you found?

In the data wrangling phase, several preprocessing and transformation steps were performed to prepare the dataset for analysis and machine learning. Initially, a copy of the original dataset was created to ensure that the raw data remains unchanged. The 'State' column was cleaned and standardized by replacing hyphens with spaces and converting text into title case to improve readability and consistency across visualizations.

The dataset was then sorted based on state, year, and quarter to maintain proper chronological order, which is essential for time-series analysis. A new feature called 'Growth' was engineered by calculating the percentage change in transaction amount for each state over consecutive time periods. This feature helps in understanding how digital transactions are evolving over time.

Since the growth calculation introduces missing values for the first occurrence of each state, these values were handled by replacing them with zero. Furthermore, a new target variable named 'Target' was created based on the growth rate. Regions with a growth rate greater than 15% were labeled as high-growth (1), while others were labeled as low-growth (0). This transformation converts the dataset into a supervised learning problem suitable for classification models.

These manipulations not only improve data quality but also enhance the dataset by introducing meaningful features that capture temporal trends and growth patterns. This plays a crucial role in enabling predictive modeling and generating actionable business insights.

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

* **Null Hypothesis (H₀):** There is no significant difference in transaction amount across different states.
* **Alternate Hypothesis (H₁):** There is a significant difference in transaction amount across different states.


#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

# Prepare data (top 5 states for simplicity)
top_states = df_clean["State"].value_counts().head(5).index

groups = [df_clean[df_clean["State"] == state]["Transaction_amount"] for state in top_states]

# ANOVA test
f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
````

##### Which statistical test have you done to obtain P-Value?

The ANOVA (Analysis of Variance) test is used to determine whether there are statistically significant differences between the means of transaction amounts across multiple states.


##### Why did you choose the specific statistical test?

ANOVA is chosen because we are comparing the means of a numerical variable (transaction amount) across more than two categorical groups (states). It is the most appropriate test for identifying whether at least one group differs significantly from the others.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Group by quarter
groups = [df_clean[df_clean["Quarter"] == q]["Transaction_amount"] for q in df_clean["Quarter"].unique()]

# ANOVA test
f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
````

##### Which statistical test have you done to obtain P-Value?

ANOVA test is used to compare transaction amounts across multiple quarters.
