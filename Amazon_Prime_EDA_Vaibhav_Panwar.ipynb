## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# 1. Remove duplicate rows
titles = titles.drop_duplicates()
credits = credits.drop_duplicates()

# 2. Handle missing values

# Fill missing IMDb scores with mean
titles['imdb_score'] = titles['imdb_score'].fillna(titles['imdb_score'].mean())

# Fill missing IMDb votes with 0
titles['imdb_votes'] = titles['imdb_votes'].fillna(0)

# Drop rows where title is missing
titles = titles.dropna(subset=['title'])

# 3. Convert data types
titles['release_year'] = pd.to_numeric(titles['release_year'], errors='coerce')
titles['runtime'] = pd.to_numeric(titles['runtime'], errors='coerce')

# 4. Clean text columns (optional formatting)
titles['genres'] = titles['genres'].astype(str)
titles['production_countries'] = titles['production_countries'].astype(str)

# 5. Reset index after cleaning
titles.reset_index(drop=True, inplace=True)
credits.reset_index(drop=True, inplace=True)

# Final check
print("Titles dataset shape after cleaning:", titles.shape)
print("Credits dataset shape after cleaning:", credits.shape)
````
